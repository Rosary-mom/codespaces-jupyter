```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt

print("NumPy version:" , np.__version__)
print("Matplotlib version" , plt.matplotlib.__version__)
print("qutip version" , qt.__version__)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Test Sinus-Kurve (z. B. f√ºr Verkehrsfluss)")
plt.xlabel("Zeit")
plt.ylabel("Wert")
plt.show()
```

```python
# Quanten-Zustand erstellen (z. B. f√ºr bin√§re Verkehrsentscheidungen: frei/besetzt)
ket = qt.basis(2, 0)
print("Quanten-Zustand:", ket)

# Bloch-Sph√§re plotten (visuelle Darstellung von Quanten-Zust√§nden)
bloch = qt.Bloch()
bloch.add_states(ket)
bloch.show()
```
